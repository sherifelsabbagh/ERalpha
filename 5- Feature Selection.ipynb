+ **Removing Multicolinearity**

Collinearity is a very common problem in machine learning projects. It is the correlation between the features of a dataset and it can reduce the performance of our models because it increases variance and the number of dimensions. It becomes worst when you have to work with unsupervised models.

```python
#!pip install collinearity
```

```python
from collinearity import SelectNonCollinear
from sklearn.feature_selection import f_classif
```

```python
def remove_collinearity(df):
    selector = SelectNonCollinear(0.9,scoring=f_classif)
    X= df.drop(["Name","label","pIC50"],axis=1)
    y = df["label"]
    selector.fit(X.values,y)
    df_ = pd.DataFrame(selector.transform(X.values),columns=X.columns[selector.get_support()])
    return df_

    
    
```

```python
df_ap_ = remove_collinearity(df_ap)
df_apc_= remove_collinearity(df_apc)
df_estate_= remove_collinearity(df_estate)
df_ext_= remove_collinearity(df_ext)
df_fp_= remove_collinearity(df_fp)
df_graph_= remove_collinearity(df_graph)
df_maccs_= remove_collinearity(df_maccs)
df_pc_= remove_collinearity(df_pc)
df_ss_= remove_collinearity(df_ss)
df_sscount_= remove_collinearity(df_sscount)
```

```python
def build_model_class(df):
    X = df.drop(["Name","label","pIC50"],axis=1)
    y = df["label"]
    ##Splitting dataset 
    scaler = MinMaxScaler()
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,shuffle=True)
    X_train = pd.DataFrame(scaler.fit_transform(X_train),columns=X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test),columns=X_test.columns)
    ##Fitting to training data
    rf= RandomForestClassifier(random_state=42)
    rf.fit(X_train,y_train)
    ## training set predicition
    y_train_pred = rf.predict(X_train)
    ## test set predictions
    y_test_pred = rf.predict(X_test)
    ## 10-Fold cross validation
    cv_results = cross_validate(rf,X_train,y_train,cv=10,scoring=["precision","recall"])
    ## Training set validation
    training_precision = precision_score(y_train,y_train_pred,average="micro")
    training_recall = recall_score(y_train,y_train_pred,average="micro")
    ## Test set validation
    test_precision = precision_score(y_test,y_test_pred,average="micro")
    test_recall = recall_score(y_test,y_test_pred,average="micro")
    return y_train_pred,y_test_pred,cv_results,training_precision,training_recall,test_precision,test_recall

```

```python
ap_results=  build_model_class(df_ap_)
apc_results=  build_model_class(df_apc_)
estate_results=  build_model_class(df_estate_)
extended_results=  build_model_class(df_ext_)
graph_results=  build_model_class(df_graph_)
fingerprint_results=  build_model_class(df_fp_)
maccs_results=  build_model_class(df_maccs_)
pubchem_results=  build_model_class(df_pc_)
substr_results=  build_model_class(df_ss_)
substrucount_results=  build_model_class(df_sscount_)
```
